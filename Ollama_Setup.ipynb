## Running Ollama
---

In order to use Ollama it needs to run as a service in background parallel to your scripts. Becasue Jupyter Notebooks is built to run code blocks in sequence this make it difficult to run two blocks at the same time. As a workaround we will create a service using subprocess in Python so it doesn't block any cell from running.

Service can be started by command `ollama serve`.

`time.sleep(5)` adds some delay to get the Ollama service up before downloading the model.

```python
!curl -fsSL https://ollama.com/install.sh | sh
```

```python
import threading
import subprocess
import time

def start_ollama():
    subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

threading.Thread(target=start_ollama, daemon=True).start()
time.sleep(5)  # 等几秒让服务起来
print("Ollama 服务已启动！")
```

```python
import subprocess
import threading
import time
import socket
import re  # 用于从日志提取 URL

def start_cloudflare_tunnel(port=11434):
    print("启动 Cloudflare Tunnel（后台模式）...")

    # 检查端口是否已开（Ollama 就绪）
    timeout = 60
    start = time.time()
    while time.time() - start < timeout:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        sock.close()
        if result == 0:
            print("Ollama 端口已就绪，开始创建隧道...")
            break
        time.sleep(1)
    else:
        print("超时：Ollama 未启动，请先运行 !ollama serve &")
        return

    # cloudflared 命令：加 --http-host-header 修复 403，加 --loglevel debug 看日志
    cmd = [
        "cloudflared", "tunnel",
        "--url", f"http://localhost:{port}",
        "--http-host-header", "localhost:11434",
        "--loglevel", "debug"
    ]

    # 用 Popen 后台启动，捕获 stdout/stderr
    process = subprocess.Popen(
        cmd,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True,
        bufsize=1,
        universal_newlines=True
    )

    # 实时读取日志，提取 URL 并打印
    url_found = False
    for line in iter(process.stdout.readline, ''):
        line = line.strip()
        if line:
            print(line)  # 打印所有日志，便于调试

        # 匹配 trycloudflare.com URL
        match = re.search(r'(https?://[^\s]+\.trycloudflare\.com)', line)
        if match and not url_found:
            public_url = match.group(1)
            print("\n" + "="*60)
            print(f"🎉 Cloudflare Tunnel 后台启动成功！公共 URL: {public_url}")
            print(f"测试模型列表: {public_url}/api/tags")
            print(f"OpenAI 兼容 Base URL: {public_url}/v1")
            print("隧道已作为后台进程运行，不会阻塞 cell。")
            print("="*60 + "\n")
            url_found = True

        # 可选：如果想在找到 URL 后继续打印日志或 break
        # if url_found: break  # 注释掉这行以持续看日志

    # 等待进程结束（实际不会结束，除非手动杀）
    process.wait()

# 用线程启动隧道（daemon=True 让它随 runtime 结束）
thread = threading.Thread(target=start_cloudflare_tunnel, daemon=True)
thread.start()

print("隧道线程已启动！你可以继续执行其他 cell。")
print("等 10-60 秒，看上面日志中是否出现 URL。")
time.sleep(60)
```
